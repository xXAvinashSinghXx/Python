```python
# Check whether its leap year or not
def is_leap(year):
    leap = False
    if (year % 4 == 0):
        if (year % 100 == 0):
            if (year % 400 == 0):
                leap = True
        else:
            leap = True

    return leap
    
year = int(input())
print(is_leap(year))
```

```python
# Check whether its leap year or not
def is_leap(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

year = int(input())
print(f"{year} is a leap year: {is_leap(year)}")
```

```python
# Simple To-Do List Program
tasks = []

def add_task(task):
    tasks.append(task)
    print(f"'{task}' has been added.")

def show_tasks():
    if not tasks:
        print("Your to-do list is empty!")
    else:
        print("Your To-Do List:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")

def remove_task(index):
    if 0 < index <= len(tasks):
        removed = tasks.pop(index - 1)
        print(f"'{removed}' has been removed.")
    else:
        print("Invalid task number.")

while True:
    print("\n1. Add Task\n2. Show Tasks\n3. Remove Task\n4. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        task = input("Enter the task: ")
        add_task(task)
    elif choice == "2":
        show_tasks()
    elif choice == "3":
        try:
            index = int(input("Enter task number to remove: "))
            remove_task(index)
        except ValueError:
            print("Please enter a valid number!")
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice, try again.")
```

```python
from itertools import product

password = list(input("Enter an 8-character password: "))

a = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*~`.'

for i in product(a, repeat=len(password)):
    if list(i) == password:
        print(f"Password found: {''.join(i)}")
        break
```
